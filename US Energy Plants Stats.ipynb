```python
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:", fig_size)
 
# Set figure width to 14 and height to 11
fig_size[0] = 14
fig_size[1] = 11
plt.rcParams["figure.figsize"] = fig_size
plants.groupby(['Primary Technology','State'])['Primary Technology'].size().unstack().plot(kind='bar',stacked=True)
plt.legend(bbox_to_anchor=(0, 1), loc='upper left', ncol=4)
plt.show()
```
